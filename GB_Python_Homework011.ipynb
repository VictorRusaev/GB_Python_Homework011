```python
import sympy
x = sympy.Symbol('x')
f = (sympy.sin(x))**2 - (sympy.cos(x))**2
```

```python
sympy.solveset(f)
```

```python
sympy.solveset(sympy.diff(f)>0)
```

```python
sympy.solveset(sympy.diff(f)<0)
```

```python
sympy.plot(f)
```

```python
sympy.solve(sympy.diff(f))
```

```python
sympy.solve(f > 0)
```

```python
sympy.solve(f < 0)
```
